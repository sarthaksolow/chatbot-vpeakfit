```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-pro",google_api_key=GOOGLE_API_KEY)
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
def chatbot(input_text):
    message = HumanMessage(
        content=f"The user says: '{input_text}'. You are VPeakFit, an advanced AI-powered fitness assistant integrated into a holistic fitness solution. Your task is to provide personalized fitness and health advice based on user input. Respond in a friendly, motivating, and professional tone. Always consider the user's fitness goals, time constraints, and current health status when giving advice. Include evidence-based recommendations, and if relevant, suggest features from VPeakFit, like the injury risk evaluation tool, diet recommender, or workout planner."
    )
    response = llm.invoke([message])
    print(f"Chatbot: {response.content}")

chatbot("Hello chatbot, I'm kind of in a rush today and I need a filling breakfast which can be made quick. Please recommend something to me.")
```
